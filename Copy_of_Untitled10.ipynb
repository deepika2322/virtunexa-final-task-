```python
import random
import logging

# Setup logging to file
logging.basicConfig(filename="history.txt", level=logging.INFO,
                    format="%(asctime)s - %(message)s")



class FlashcardApp:
    def __init__(self):
        self.flashcards = []

    def add_flashcard(self):
        question = input("Enter the question: ")
        answer = input("Enter the answer: ")
        self.flashcards.append({"question": question, "answer": answer})
        print("‚úÖ Flashcard added successfully!\n")

    def quiz(self):
        if not self.flashcards:
            print("‚ö†Ô∏è No flashcards available. Add some first.\n")
            return

        print("\nüß† Quiz Time!")
        score = 0
        random.shuffle(self.flashcards)

        for card in self.flashcards:
            user_answer = input(f"Q: {card['question']} > ").strip().lower()
            correct_answer = card['answer'].strip().lower()
            if user_answer == correct_answer:
                print("‚úÖ Correct!\n")
                score += 1
            else:
                print(f"‚ùå Incorrect. The correct answer is: {card['answer']}\n")

        print(f"üéØ Quiz Finished! Your Score: {score}/{len(self.flashcards)}\n")



class CalculatorApp:
    def calculate(self):
        try:
            a = float(input("Enter first number: "))
            op = input("Enter operation (+, -, *, /): ")
            b = float(input("Enter second number: "))

            if op == '+':
                result = a + b
            elif op == '-':
                result = a - b
            elif op == '*':
                result = a * b
            elif op == '/':
                if b == 0:
                    print("‚ö†Ô∏è Error: Division by zero!\n")
                    return
                result = a / b
            else:
                print("‚ö†Ô∏è Invalid operator!\n")
                return

            print(f"‚úÖ Result: {result}\n")
            logging.info(f"CALC: {a} {op} {b} = {result}")

        except ValueError:
            print("‚ö†Ô∏è Error: Invalid input! Please enter numeric values.\n")



def flashcard_menu(flashcard_app):
    while True:
        print("üìö Flashcard Menu:\n1. Add Flashcard\n2. Take Quiz\n3. Back to Main Menu")
        choice = input("Enter your choice: ")
        if choice == '1':
            flashcard_app.add_flashcard()
        elif choice == '2':
            flashcard_app.quiz()
        elif choice == '3':
            break
        else:
            print("‚ö†Ô∏è Invalid choice. Please try again.\n")

def calculator_menu(calculator_app):
    while True:
        print("üßÆ Calculator Menu:\n1. Perform Calculation\n2. Back to Main Menu")
        choice = input("Enter your choice: ")
        if choice == '1':
            calculator_app.calculate()
        elif choice == '2':
            break
        else:
            print("‚ö†Ô∏è Invalid choice. Please try again.\n")

def main():
    flashcard_app = FlashcardApp()
    calculator_app = CalculatorApp()

    while True:
        print("üìå Main Menu:\n1. Flashcards\n2. Calculator\n3. Exit")
        choice = input("Enter your choice: ")
        if choice == '1':
            flashcard_menu(flashcard_app)
        elif choice == '2':
            calculator_menu(calculator_app)
        elif choice == '3':
            print("üëã Exiting... Goodbye!")
            break
        else:
            print("‚ö†Ô∏è Invalid choice. Please try again.\n")



if __name__ == "__main__":
    main()
```
